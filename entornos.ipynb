```python
# Cargar el Conjunto de Datos
import pandas as pd

# URL del conjunto de datos "Online Retail" del repositorio UCI
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00352/Online%20Retail.xlsx"

# Se usa read_excel ya que la fuente es un archivo Excel

data = pd.read_excel(url)

print(data.head())
```

```python
# Realizar una Exploracion Basica de Datos
import matplotlib.pyplot as plt
import seaborn as sns
print("Valores faltantes por columna:\n", data.isnull().sum())
print("\nEstadısticas descriptivas:\n", data.describe())
data = data.dropna(subset=['CustomerID']) # Paso adicional: eliminar filas sin cliente
data = data[data['Quantity'] > 0]
plt.figure(figsize=(10, 8))
sns.countplot(y='Country', data=data, order=data['Country'].value_counts().index[:15])
plt.title('Top 15 de Paıses por N´umero de Transacciones')
plt.xlabel('Numero de Transacciones')
plt.ylabel('Paıs')
plt.show()
```

```python
# Analisis de Series de Tiempo con statsmodels
import statsmodels.api as sm

# Preparar los datos: usar InvoiceDate como ´ındice y agregar ventas diarias
data['InvoiceDate'] = pd.to_datetime(data['InvoiceDate'])
daily_sales = data.set_index('InvoiceDate')['Quantity'].resample('D').sum()

# Realizar la descomposici´on estacional
decomposition = sm.tsa.seasonal_decompose(daily_sales.dropna(), model='additive')

# Visualizar los componentes
fig = decomposition.plot()
plt.show()
```

```python
# Visualizacion Declarativa con altair
# Limpieza básica
data = data.dropna(subset=['Quantity', 'UnitPrice'])
data = data[(data['Quantity'] > 0) & (data['UnitPrice'] > 0)]
data['InvoiceNo'] = data['InvoiceNo'].astype(str)  # evita error de conversión

# Habilitar vegafusion (para datasets grandes) ---
import altair as alt
alt.data_transformers.enable("vegafusion")

# Crear gráfico Altair
scatter_chart = (
    alt.Chart(data)
    .mark_circle(size=60, color='steelblue', opacity=0.6)
    .encode(
        x=alt.X('Quantity:Q', title='Cantidad'),
        y=alt.Y('UnitPrice:Q', title='Precio Unitario'),
        tooltip=['InvoiceNo:N', 'StockCode:N', 'Quantity:Q', 'UnitPrice:Q', 'Country:N']
    )
    .properties(
        title='Gráfico de dispersión: Cantidad vs. Precio Unitario (Online Retail)',
        width=600,
        height=400
    )
    .interactive()
)

scatter_chart
```

```python
import pandas as pd
from dash import Dash, dcc, html, Input, Output
import plotly.express as px

# --- Cargar y limpiar datos ---
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00352/Online%20Retail.xlsx"
data = pd.read_excel(url)

data = data.dropna(subset=['Quantity','UnitPrice','Country','Description'])
data = data[(data['Quantity']>0) & (data['UnitPrice']>0)]

paises = sorted(data['Country'].unique())

# Crear app Dash
app = Dash(__name__)

app.layout = html.Div([
    html.H1("Top 10 productos más vendidos por país"),
    dcc.Dropdown(
        id='dropdown-pais',
        options=[{'label': pais, 'value': pais} for pais in paises],
        value='United Kingdom',
        clearable=False,
        style={'width':'50%'}
    ),
    dcc.Graph(id='grafico-productos')
])

@app.callback(
    Output('grafico-productos','figure'),
    Input('dropdown-pais','value')
)
def actualizar_grafico(pais):
    df = data[data['Country']==pais]
    top = df.groupby('Description')['Quantity'].sum().sort_values(ascending=False).head(10).reset_index()
    fig = px.bar(top, x='Description', y='Quantity', color='Quantity', color_continuous_scale='Blues')
    fig.update_layout(xaxis_tickangle=-45)
    return fig

# --- Ejecutar servidor ---
if __name__ == '__main__':
    # Esto abrirá la app en un navegador, no dentro de la celda
    app.run(debug=True)
```
